```python
!pip install sympy
import numpy as np
import sympy as sp

print("--- Operaciones con Matrices ---")

A = np.array([[2, 1],
              [0, 3]])

B = np.array([[1, 2],
              [3, 0]])

print(f"Matriz A:\n{A}")
print(f"Matriz B:\n{B}\n")

suma_AB = A + B
print(f"Suma A + B:\n{suma_AB}\n")

producto_AB = A @ B  
print(f"Producto A x B:\n{producto_AB}\n")

try:
    inversa_A = np.linalg.inv(A)
    print(f"Inversa de A (A^-1):\n{inversa_A}\n")
except np.linalg.LinAlgError:
    print("La matriz A no tiene inversa (es singular).\n")

det_B = np.linalg.det(B)
print(f"Determinante de B: {det_B}\n")


print("--- Operaciones con Vectores ---")

u = np.array([1, 3])
v = np.array([-1, 2])

print(f"Vector u: {u}")
print(f"Vector v: {v}\n")

producto_interno_uv = np.dot(u, v) 
print(f"Producto interno u¬∑v: {producto_interno_uv}\n")

magnitud_u = np.linalg.norm(u)
magnitud_v = np.linalg.norm(v)
print(f"Magnitud de u (||u||): {magnitud_u}")
print(f"Magnitud de v (||v||): {magnitud_v}\n")


print("--- C√°lculo de Derivadas ---")

x = sp.symbols('x')
h = sp.sin(2*x**2 + 3*x)
print(f"Funci√≥n h(x) = {h}\n")

h_prima = sp.diff(h, x)
print(f"Derivada h'(x) = {h_prima}\n")

valor_h_prima_1 = h_prima.subs(x, 1)
print(f"h'(1) = {valor_h_prima_1}")
print(f"Valor num√©rico de h'(1) ‚âà {valor_h_prima_1.evalf()}\n") 


valor_h_prima_0 = h_prima.subs(x, 0)
print(f"h'(0) = {valor_h_prima_0}")
print(f"Valor num√©rico de h'(0) = {valor_h_prima_0.evalf()}\n") 
```
