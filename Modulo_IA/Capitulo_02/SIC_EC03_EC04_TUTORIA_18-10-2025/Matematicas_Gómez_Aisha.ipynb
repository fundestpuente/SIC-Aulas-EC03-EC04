```python
import numpy as np
import sympy as sp

## Ejercicio 5

# Matrices

A = [[1, 1], [2, 3]]
B = [[0, 2], [1, 1]]
A = np.array(A)
B = np.array(B)

# Calcula A + B, A - B y A √ó B.

print("Suma de matrices (A + B):")
C = A + B
print(C)

print("Resta de matrices (A - B):")
D = A - B
print(D)

print("Producto matricial (A √ó B):")
E = np.dot(A, B)
print(E)


# Calcula la traspuesta de A y la inversa de B (si existe).

print("Traspuesta de A:")
A_tras = np.transpose(A)
print(A_tras)

print("Inversa de B (si existe):")
try:
    B_inv = np.linalg.inv(B)
    print(B_inv)
except np.linalg.LinAlgError:
    print("La matriz B no tiene inversa (es singular).")
# Vectores

u = [2, 3]
v = [1, -1]
u = np.array(u)
v = np.array(v)


# Calcula el producto interno u¬∑v y determina ortogonalidad.

print("Producto interno u¬∑v:")
prod_int = np.dot(u, v)
print(prod_int)

if prod_int == 0:
    print("Los vectores son ortogonales.")
else:
    print("Los vectores no son ortogonales.")

# Calcula magnitudes y √°ngulo entre u y v.

print("Magnitud de u:")
mag_u = np.linalg.norm(u)
print(mag_u)

print("Magnitud de v:")
mag_v = np.linalg.norm(v)
print(mag_v)

print("√Ångulo entre u y v:")
ang = np.arccos(np.dot(u, v) / (mag_u * mag_v))
print(ang)

# œÉ(x) = 1 / (1 + exp(-x))  # Funci√≥n sigmoide

X = sp.symbols('x')
sig = 1 / (1 + sp.exp(-X))
print("œÉ(x) =")
print(sig)

# Calcula œÉ'(x) simb√≥licamente y simplif√≠cala.

print("Derivada simb√≥lica œÉ'(x):")
dev = sp.diff(sig, X)
print(dev)

print("Derivada simplificada œÉ'(x):")
dev_simp = sp.simplify(dev)
print(dev_simp)

# Eval√∫a œÉ'(0) y œÉ'(1) y explica su interpretaci√≥n.

print("Evaluaci√≥n de œÉ'(x) en x = 0 y x = 1:")
sigmoide_0 = dev_simp.subs(X, 0)
sigmoide_1 = dev_simp.subs(X, 1)

print("œÉ'(0) =", sigmoide_0)
print("œÉ'(1) =", sigmoide_1)


# Interpretaci√≥n:

# La derivada es m√°xima en x = 0, lo que significa que la funci√≥n sigmoide cambia
# m√°s r√°pido cerca del origen. Aqu√≠ la funci√≥n es m√°s sensible a cambios en X, mientras que
# si se aleja de 0, como en 1, la derivada disminuye, y eso muestra que la funci√≥n comienza
# a saturarse y los cambios realizados en X no producen grandes cambios en Y.
```
