```python
import sympy as sp


"""
EJERCICIO 4

Matrices:
A = [[1, 2, 0], [0, 1, 1], [3, 0, 2]]
B = [[2, 1, 1], [1, 0, 3], [0, 2, 1]]

Calcula A × B y la traspuesta de B.
Calcula el determinante de A y, si existe, su inversa.


Vectores:
u = [1, -2, 3]
v = [0, 1, 4]

Calcula el producto interno u·v.
Calcula la magnitud y el ángulo entre los vectores.


Derivadas:
f(x) = x4 - 3*x3 + 2*x - 5

Calcula f'(x) y f''(x).
Evalúa en x = 2 y comenta la concavidad.


Nombre del estudiante: Diego Montesdeoca
"""

print("="*50)
print("MATRICES")
A = [[1, 2, 0],
     [0, 1, 1],
     [3, 0, 2]]

B = [[2, 1, 1],
     [1, 0, 3],
     [0, 2, 1]]

matrizA = sp.Matrix(A)
matrizB = sp.Matrix(B)

print("Producto de A x B: ", matrizA * matrizB)

print("Matriz transpuesta de B:", matrizB.transpose())

detA = matrizA.det()
if (detA != 0):
    print("Determinante de A:", detA)
    print("Matriz inversa de A:", matrizA.inv())
else:
    print("La matriz no tiene Inversa, por su determinante:", detA)

print("="*50)
print("VECTORES")
u = [1, -2, 3]
v = [0, 1, 4]

vectorU = sp.Matrix(u)
vectorV = sp.Matrix(v)

print("Producto interno u.v:", vectorU.dot(vectorV))

print("Magnitud entre u y v:", (vectorV-vectorU).norm())

#Angulo usando producto punto:   cos(Angulo_entre_vectores) * magnitud(A) * magnitud(B) = magnitud(producto_punto(A, B))
#                               Angulo_entre_vectores = cos^-1(magnitud(producto_punto(A, B)) / magnitud(A) * magnitud(B))

angulo_entre_vectores = sp.acos((vectorV.dot(vectorU)) / (vectorV.norm() * vectorU.norm()))
print("Angulo entre V y U en radianes:", angulo_entre_vectores.round(2))
print("Angulo entre V y U en grados:", sp.deg(angulo_entre_vectores).round(2))

print("="*50)
print("DERIVADAS")

x = sp.symbols('x')
f = x**4 - 3*x**3 + 2*x - 5

dev1_f = sp.diff(f)
dev2_f = sp.diff(sp.diff(f))

print("La primera derivada de f(x) = x4 - 3*x3 + 2*x - 5:", dev1_f)
print("La segunda derivada:", dev2_f)

dev2_f_2 = dev2_f.subs(x, 2)
print("La segunda derivada evaluado en 2:", dev2_f_2)
print("f(x) es concavo", "hacia arriba" if (dev2_f_2 > 0) else "hacia abajo", "en x = 2")
```
